### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Agrupar por la columna 'Tienda' y calcular el ingreso total
ingreso_tienda = sum(tienda["Precio"])
print("Ingreso total Tienda 1:", ingreso_tienda)
ingreso_tienda2 = sum(tienda2["Precio"])
print("Ingreso total Tienda 2:", ingreso_tienda2)
ingreso_tienda3 = sum(tienda3["Precio"])
print("Ingreso total Tienda 3:", ingreso_tienda3)
ingreso_tienda4 = sum(tienda4["Precio"])
print("Ingreso total Tienda 4:", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
categorias = tienda["Categoría del Producto"].value_counts() # Conocer las categorias de la tienda 1
print(f"Tienda 1 {categorias}")

categorias_tienda_2 = tienda2["Categoría del Producto"].value_counts() # Conocer las categorias de la tienda 2
print(f"Tienda 2 {categorias_tienda_2}")

categorias_tienda_3 = tienda3["Categoría del Producto"].value_counts() # Conocer las categorias de la tienda 3
print(f"Tienda 3 {categorias_tienda_3}")

categorias_tienda_4 = tienda4["Categoría del Producto"].value_counts() # Conocer las categorias de la tienda 4
print(f"Tienda 4 {categorias_tienda_4}")
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda = (tienda["Calificación"].mean())
print("Calificación promedio Tienda 1:", calificacion_tienda)

calificacion_tienda_2= (tienda2["Calificación"].mean())
print("Calificación promedio Tienda 2:", calificacion_tienda_2)

calificacion_tienda_3= (tienda3["Calificación"].mean())
print("Calificación promedio Tienda 3:", calificacion_tienda_3)

calificacion_tienda_4= (tienda4["Calificación"].mean())
print("Calificación promedio Tienda 4:", calificacion_tienda_4)
```

# 4. Productos más y menos vendidos

```python
Producto_mas_vendido = tienda["Producto"].value_counts().idxmax()
print("Producto más vendido Tienda 1:", Producto_mas_vendido)

Producto_menos_vendido = tienda["Producto"].value_counts().idxmin()
print("Producto menos vendido Tienda 1:", Producto_menos_vendido)

Producto_mas_vendido_2 = tienda2["Producto"].value_counts().idxmax()
print("Producto más vendido Tienda 2:", Producto_mas_vendido_2)

Producto_menos_vendido_2 = tienda2["Producto"].value_counts().idxmin()
print("Producto menos vendido Tienda 2:", Producto_menos_vendido_2)

Producto_mas_vendido_3 = tienda3["Producto"].value_counts().idxmax()
print("Producto más vendido Tienda 3:", Producto_mas_vendido_3)

Producto_menos_vendido_3 = tienda3["Producto"].value_counts().idxmin()
print("Producto menos vendido Tienda 3:", Producto_menos_vendido_3)

Producto_mas_vendido_4 = tienda4["Producto"].value_counts().idxmax()
print("Producto más vendido Tienda 4:", Producto_mas_vendido_4)

Producto_menos_vendido_4 = tienda4["Producto"].value_counts().idxmin()
print("Producto menos vendido Tienda 4:", Producto_menos_vendido_4)
```

# 5. Envío promedio por tienda

```python
Costo_envio_tienda1 = (tienda["Costo de envío"].mean())
print(f"El costo de envío de la tienda 1 es: {Costo_envio_tienda1}")

Costo_envio_tienda2 = (tienda2["Costo de envío"].mean())
print(f"El costo de envío de la tienda 2 es: {Costo_envio_tienda2}")

Costo_envio_tienda3 = (tienda3["Costo de envío"].mean())
print(f"El costo de envío de la tienda 3 es: {Costo_envio_tienda3}")

Costo_envio_tienda4 = (tienda4["Costo de envío"].mean())
print(f"El costo de envío de la tienda 4 es: {Costo_envio_tienda4}")

menor_costo_envio=min(Costo_envio_tienda1,Costo_envio_tienda2,Costo_envio_tienda3,Costo_envio_tienda4)
print(f"El menor costo de envío es: {menor_costo_envio}")

mayor_costo_envio=max(Costo_envio_tienda1,Costo_envio_tienda2,Costo_envio_tienda3,Costo_envio_tienda4)
print(f"El mayor costo de envío es: {mayor_costo_envio}")
```
